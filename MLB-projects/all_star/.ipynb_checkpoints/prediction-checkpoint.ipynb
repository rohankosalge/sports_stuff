Import Modules

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from unidecode import unidecode # if needed to remove accents
```

Gather Data into Train/Test Split

```python
start_train_year, end_train_year, test_year = 2016, 2020, 2019
normalize = True
```

```python
X_train, y_train = pd.DataFrame(), pd.DataFrame()
scaler = StandardScaler()
for year in range(start_train_year, end_train_year+1):
    if year == test_year:
        continue
    train_year = pd.read_csv('/Users/rohankosalge/all_star/data/as_'+str(year)+'.csv')
    X_train_year, y_train_year = train_year[train_year.columns[2:-1]], train_year[train_year.columns[-1]]
    if normalize:
        X_train = pd.concat([X_train, pd.DataFrame(scaler.fit_transform(X_train_year))])
    else:
        X_train = pd.concat([X_train, X_train_year])
    y_train = pd.concat([y_train, y_train_year])
```

```python
test = pd.read_csv('/Users/rohankosalge/all_star/data/as_'+str(test_year)+'.csv')
if normalize:
    X_test = pd.DataFrame(scaler.fit_transform(test[test.columns[2:-1]]))
else:
    X_test = test[test.columns[2:-1]].rename_axis(None, axis=1)
y_test = test[test.columns[-1]]
labels = test[test.columns[0]]
```

Create and Fit Model

```python
model = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
outputs = model.predict_proba(X_test)
outputs
```

```python
outputs_df = pd.DataFrame(outputs[:,1])
print(len(outputs_df.index), len(y_test.index))
outputs_df = pd.concat([labels, outputs_df, y_test], axis=1).sort_values(by=0, ascending=False)
outputs_df.rename(columns={'Unnamed: 0': 'Name', 0: 'Probability'}, inplace=True)
outputs_df['Probability'] = outputs_df['Probability'].apply(lambda x: "{:.2f}".format(float(x)))
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(outputs_df)
```
