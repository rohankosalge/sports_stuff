```python
start_year, end_year = 2016, 2020
```

```python
p = PageviewsClient(user_agent="Python query script by " + contact_info)
```

```python
for year in range(start_year, end_year+1):
    schedule = pd.read_csv('/Users/rohankosalge/Documents/old-mac-code/harp-code/NBA-Stats-HARP-Project/Schedule-Stats/NBARegularSeasonGames-'+str(year-1)+'_'+str(year)+'.csv')
    schedule['Game Date'] = schedule['Game Date'].str[4:]
    date = reg_date + ' ' + str(year)
    schedule = schedule.loc[:schedule[schedule['Game Date']==date].last_valid_index(),:]
    with open('/Users/rohankosalge/Desktop/Coding/NBA-projects/player-evaluations-proj/player-evals/'+str(year)+'-player-evals.json', 'r') as json_file:
        games = json.load(json_file)[:len(schedule.index)]
    players = {}
    for game in games:
        tm1, tm2 = game[0][:3], game[0][4:]
        for tm in [tm1, tm2]:
            for perf in game[1][tm]:
                player, score = perf[0], perf[1]
                if player in players:
                    if tm != players[player][0]:
                        players[player][0] = tm
                    players[player].append(score)
                else:
                    players[player] = [tm, score]

    with open('/Users/rohankosalge/Desktop/Coding/NBA-projects/all-star-pred/all_star_appearances.json', 'r') as json_file:
        prev = json.load(json_file)
    
    for player in players:
        if len(players[player]) == 2:
            players[player].append(round(sum(players[player][1:])/len(players[player][1:]), 1))
            players[player][1] = 1
        else:
            players[player][2] = round(sum(players[player][1:])/len(players[player][1:]), 1)
            players[player][1] = len(players[player][1:])
        players[player] = players[player][:3]

        team = init_to_tm[players[player][0]]

        index1 = schedule['Away'].str.contains(team)[schedule['Away'].str.contains(team)].index[-1]
        index2 = schedule['Home'].str.contains(team)[schedule['Home'].str.contains(team)].index[-1]
        if index1 > index2:
            record = schedule.iloc[index1]['Away Record'].split('-')
        else:
            record = schedule.iloc[index2]['Home Record'].split('-')

        players[player][0] = round(int(record[0])/(int(record[0])+int(record[1])), 3)

        if player in prev:
            prev_as = prev[player]
            players[player].append(len([pyear for pyear in prev_as if pyear < year]))
            if year in prev_as:
                players[player].append(1)
            else:
                players[player].append(0)
        else:
            players[player].append(0)
            players[player].append(0)


    players_df = pd.DataFrame.from_dict(players).T
    players_df = players_df.rename(columns={0: 'Win%', 1: 'Games', 2: 'Val', 3: 'PrevAS', 4: 'AS'})

    input_players = [player.replace('J.J.', 'J. J.').replace('J.R.', 'J. R.').replace('A.J.', 'A. J.').replace('C.J.', 'C. J.').replace('J. J. Redick', 'JJ Redick') for player in list(players.keys())]
    views = p.article_views(project='en.wikipedia',
                            articles=input_players,
                            granularity='daily',
                            start=str(year-1)+'110100',
                            end=str(year)+'012100')

    views_df = pd.DataFrame(views).agg("sum", axis="columns").to_frame()
    views_df.index = views_df.index.str.replace('_', ' ')
    views_df = views_df.fillna(0) 
    #print(views_df)
    players_df.index = players_df.index.str.replace('J.J.', 'J. J.')
    players_df.index = players_df.index.str.replace('J.R.', 'J. R.')
    players_df.index = players_df.index.str.replace('A.J.', 'A. J.')
    players_df.index = players_df.index.str.replace('C.J.', 'C. J.')
    players_df.index = players_df.index.str.replace('J. J. Redick', 'JJ Redick')
    players_df = pd.concat([players_df, views_df], axis=1)
    players_df = players_df.rename(columns={0:'WikiViews'})
    players_df['Year'] = year

    new_cols = ['Year', 'Win%', 'Games', 'Val', 'PrevAS', 'WikiViews', 'AS']
    players_df = players_df[new_cols].sort_values(by=['AS'], ascending=False)

    #print(players_df)
    players_df.to_csv('/Users/rohankosalge/all_star/data/as_'+str(year)+'.csv')

print('done')
players_df
    
```
