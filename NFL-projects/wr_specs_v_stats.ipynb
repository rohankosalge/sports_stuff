```python
import pandas as pd
```

```python
combine_stats = pd.read_csv('nfl_combine_since_2010.csv')
combine_stats
```

```python
wr_combine_stats = combine_stats[combine_stats['Pos'] == 'WR']
wr_combine_stats
```

```python
receiving_stats_2023 = pd.read_csv('receivers_2023.csv')
wrs_2023 = receiving_stats_2023[receiving_stats_2023['Pos'] == 'WR']
wrs_2023['Player'] = wrs_2023['Player'].map(lambda x: x.rstrip('+').rstrip('*'))
wrs_2023
```

```python
wrs_2023 = pd.merge(combine_stats, wrs_2023, how='inner', on=['Player'])
wrs_2023 = wrs_2023.sort_values(by=['Yds'], ascending=False)
wrs_2023
```

```python
forty_yd_wrs = wrs_2023[~wrs_2023['40yd'].isna()][~wrs_2023['Ht'].isna()][~wrs_2023['Wt'].isna()]
forty_yd_wrs
```

```python
def parse_ht(ht):
    # format: X-XX
    ht_ = ht.split("-")
    ft_, in_ = float(ht_[0]), float(ht_[1])
    return (12*ft_) + in_
```

```python
forty_yd_wrs['Ht'] = forty_yd_wrs['Ht'].apply(lambda x:parse_ht(x))
forty_yd_wrs['YAC%'] = forty_yd_wrs['YAC']/forty_yd_wrs['Yds']
forty_yd_wrs
```

```python
forty_yd_wrs['BMI'] = 703 * (forty_yd_wrs['Wt'] / (forty_yd_wrs['Ht']*forty_yd_wrs['Ht']))
forty_yd_wrs
```

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
plt.figure(figsize=(12,8))
plt.scatter(forty_yd_wrs['BMI'], forty_yd_wrs['40yd'])

matplotlib.rcParams['font.size'] = 7
for i, txt in enumerate(forty_yd_wrs['Player']):
    plt.annotate(txt, (forty_yd_wrs['BMI'].iloc[i], forty_yd_wrs['40yd'].iloc[i]))

plt.xlabel('BMI', fontdict={'family': 'sans-serif', 'weight': 'normal', 'size': 10})
plt.ylabel('40yd', fontdict={'family': 'sans-serif', 'weight': 'normal', 'size': 10})
plt.title('Comparing Size v. Speed in NFL WRs, 2023 Season', fontdict={'family': 'sans-serif', 'weight': 'normal', 'size': 12})
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(forty_yd_wrs['40yd'], forty_yd_wrs['YAC%'])

matplotlib.rcParams['font.size'] = 7
for i, txt in enumerate(forty_yd_wrs['Player']):
    plt.annotate(txt, (forty_yd_wrs['40yd'].iloc[i], forty_yd_wrs['YAC%'].iloc[i]))

plt.xlabel('40yd', fontdict={'family': 'sans-serif', 'weight': 'normal', 'size': 10})
plt.ylabel('YAC/Yds', fontdict={'family': 'sans-serif', 'weight': 'normal', 'size': 10})
plt.title('Comparing Speed v. YAC, WRs 2023', fontdict={'family': 'sans-serif', 'weight': 'normal', 'size': 12})
plt.show()
```
